# Prelim Activity 02


### Import Necessary Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Load the Dataset


```python
df = pd.read_excel("Prelim Activity 02.xlsx", sheet_name="LIVESTOCK.csv")
df
```

### Rename Identifier Column


```python
if "Column 1" in df.columns:
  df = df.rename(columns={"Column 1": "Record_ID"})
```

### Select Numeric Columns

```python
num_df = df.select_dtypes(include="number")
num_df = num_df.drop(columns=["Record_ID"], errors="ignore")
```

### Compute for the Mean

```python
mean = num_df.mean()
mean = mean.to_frame(name="Mean")
mean
```

### Compute for the Median

```python
median = num_df.median()
median = median.to_frame(name="Median")
median
```

### Compute for the Mode

```python
mode = num_df.mode()
mode
```

## Measure of Position

### Compute for the Quartiles

```python
# Compute for quartiles
q1 = num_df.quantile(0.25)
q3 = num_df.quantile(0.75)

quartiles = pd.DataFrame({
    "Q1 (25th Percentile)": q1,
    "Q3 (75th Percentile)": q3
}).round(2)

quartiles
```

### Compute for the Percentiles

```python
# Compute for percentiles
p5 = num_df.quantile(0.05)
p95 = num_df.quantile(0.95)

percentile = pd.DataFrame({
    "5th Percentile": p5,
    "95th Percentile": p95
}).round(2)

percentile
```

## Measure of Relative Location

### Compute for the Variance

```python
variance = num_df.var()
variance = variance.to_frame(name="Variance")
variance
```

### Compute for the Standard Deviation

```python
std_dev = num_df.std()
std_dev = std_dev.to_frame(name="Standard Deviation")
std_dev
```

## Data Visualization

### Histogram of Livestock Production (Head per Year)


```python
plt.figure(figsize=(6,4))
plt.hist(
    num_df["Livestock Production Heads Per Year"],
    bins=3,
    range=(0,300),
    edgecolor="white"
)
plt.title("Distribution of Livestock Production (Head per Year)")
plt.ylabel("Frequency")
plt.xlabel("Livestock Production (Heads per Year)")
plt.grid(axis="y", alpha=0.3)
```

### Boxplot of Livestock Production (Heads per Year)

```python
plt.figure()
plt.boxplot(
  num_df["Livestock Production Heads Per Year"],
  vert=True
)
plt.title("Boxplot of Livestock Production (Heads per Year)")
plt.ylabel("Livestock Production (Heads per Year)")
plt.tight_layout()

```
